#  Bi 1: To Bin 
*<strong> Abstraction </strong>: 
    - Nhp vo hai s nguyn n, k vi k l s phn t m n to ra tha mn iu kin ng vi mi gi tr th k s bng tng cc tch  ca cc phn t trc nhn vi hiu ch s tnh t k n cc phn t  v gi tr ca phn t u tin chnh l n; sau  tnh   tng tt c cc gi tr ca cc phn t ri in ra mn hnh.
* <strong> Pattern Recognition </strong>:
    Di s m t ca bi ton, ta c:
    - Ngy 1: sinh ra 3*1 c th mc 1 => 3 c th sinh ra 
    - Ngy 2: sinh ra 3*2 c th mc 1, 3*1 c th mc 1 => 9 c th sinh ra
    - Ngy 3: sinh ra 3*3 c th mc 1, 3*2 c th mc 1, 9*1 c th => 24 c th sinh ra
    Nu th hin di mng ta s thy s c th sinh ra tng ngy:

    |ngy|1|2 |3 |
    |------|-|- |- |
    |SL tng   |3|9 |24|
    - S c th sinh ra trong ngy th k s bng:
 a[i] = $\sum_{n = 0}^{k-1} a[i]*(k-i)$
    - Tuy nhin nhn vo tng th tng ngy ta nhn thy s khc bit 
    - Tng s c th ngy th k = tng s c th ngy (k-1)*2  + s c th tng thm trong ngy (k-1)
    
    |Ngy|0|1|2|3|
    |-|-|-|-|-|
    |Tng|3|6|15|39|
    
 - Nh vy, ta s c 1 vng lp t bt u t ch s 1, v bin s (s = 0) tnh s c th sinh ra, bin a = n l s c th ngy 0 v bin b l tng tt c cc c th theo cng thc: b=2*a+s
* <strong> K thut c p dng </strong>: K thut vng lp vi iu kin  u
* <strong> Lu  thut ton </strong>:

* <strong>  phc tp thut ton </strong>: O(n)

```python
import io, os, sys
input = sys.stdin.readline
n, k = map(int, input().split())
tong =0
a = n
for i in range(1,k+1):
    b = a*2 +tong
    tong += a
    a = b
sys.stdout.write(str(b%1000000007))
```

# Bi 2: Trm Thu Ph
* <strong> Abstraction: </strong>
- Nhp vo s lng phn t ca mt mng v mng  trn mt dng , tm khong trn mng  sao cho tng ln nht.
* <strong> K thut c p dng: </strong> Duyt mng, gn bin tm
- u tin ta gn s = 0  tnh tng, max = arr[0]
- for (i ->n), s = max (arr [i], arr[i] + s); nu s < max th max = s. Tm c ch s cui ca khong
- Tip ta li duyt ngc t v tr cui  tm c v thc hin tr ngc cc phn t trong mng cho n khi arr [i-1] = 0 hay arr [i-2] = 0 th ta s tm c ch s u 
* <strong> Lu  thut ton: </strong>

* <strong>  phc tp: </strong> O(n)

```python
n = int(input())
lst = list(map(int, input().split()))[:n]
max_lst = lst[0]
sum = 0
for i in range(n):
    sum = max(lst[i], sum + lst[i])
    if (max_lst < sum ):
        max_lst = sum
        b = i
temp = max_lst
for i in range(-(b),1) :
    temp -= lst[abs(i)]
    if (temp == 0):
        a = abs(i)
        while (lst[a-1] == 0 or (lst[a-2] + lst[a - 1] == 0)):
            if lst[a-1] == 0 :
                a -= 1
            if lst[a-2] + lst[a - 1] == 0 :
                a -= 2
        break
print (str(a+1)+" "+str(b+1)+" "+str(max_lst))
```
